```python
stopimbalance = stopimbalance.drop(columns=["Unnamed: 0"])
```

```python
# importing libraries
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
y1= stopimbalance.is_fraud[:]
x1=stopimbalance.drop("is_fraud",1)
```

```python
logreg = LogisticRegression()
rfe = RFE(estimator = logreg, n_features_to_select= 22)
rfe_res = rfe.fit(x1,y1)
```

```python
x1.columns[rfe_res.support_]
```

```python
list(zip(x1.columns,rfe.support_,rfe.ranking_))
```

```python
cols = x1.columns[rfe_res.support_]
```

```python
# making model again
model1 = sm.GLM(y,sm.add_constant(x1[cols]),family = sm.families.Binomial())
resul1 = model1.fit()
resul1.summary()
```

```python
cols = cols.drop(["recency_segment_After 6 hours","location_Nearby"],1)
```

```python
model2 = sm.GLM(y,sm.add_constant(x1[cols]),family = sm.families.Binomial())
res3 = model2.fit()
res3.summary()
```

```python
pred = res3.predict(sm.add_constant(x[cols]))
```

```python
pred
```

```python
predictions = pd.DataFrame({"Actual": y,"Pred_Prob" : pred})
```

```python
predictions.to_csv('imbalancesolution.csv')
```

```python
target_test = data_test["is_fraud"]
x_test = data_test.drop("is_fraud",1)
```

```python
x_test[["amt","Time","displacement","recency"]] = sc.fit_transform(x_test[["amt","Time","displacement","recency"]])
```

```python
result = res3.predict(sm.add_constant(x_test[cols]))
predictions_test = pd.DataFrame({"Actual": target_test,"Pred_Prob" : result})
predictions_test["Final"] = predictions_test["Pred_Prob"].map(lambda x: 0 if x<0.01 else 1)
```

```python
predictions_test
```

```python
predictions_test.to_csv('prediction_test.csv')
```

```python
import sklearn.metrics as metrics
metrics.precision_score(predictions_test["Actual"],predictions_test["Final"]),metrics.recall_score(predictions_test["Actual"],predictions_test["Final"])
```

```python
#Since there is a huge class imbalance, 
#we are encountering a huge fluctuation in precision and recall values.

#Our main motive is to identify fraud transactions 
#and build a model with high specificity
#Since, high recall model is very important as we 
#need to predict as much as fraud transactions because 
#company have to refund a full amount of fraud to the customer.
```

```python
finalaccuracy=metrics.classification_report(predictions_test["Actual"],predictions_test["Final"])
```
